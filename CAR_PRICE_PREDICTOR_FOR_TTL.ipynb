```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('car data.xls')
```

```python
import datetime
date_time = datetime.datetime.now()
data['Age']=date_time.year - data['Year']
data.drop('Year',axis=1,inplace=True)
#replacing year of purchasing with age of car
```

```python
plt.plot(data['Age'])
```

```python
plt.bar(data['Fuel_Type'], data['Age'])
```

```python
sns.boxplot(data['Selling_Price'])
# outlier check
```

```python
sns.lmplot(x = 'Kms_Driven', y = 'Selling_Price', data = data).set(title="Regression Line")
#Regression Line
```

```python
sorted(data['Selling_Price'],reverse=True)
```

```python
data = data[~(data['Selling_Price']>=33.0) & (data['Selling_Price']<=35.0)]
data.shape
# outlier removal
```

```python
sorted(data['Selling_Price'],reverse=True)
#new data check
```

# ONE-HOT ENCODING ALL X VALUES

```python
data['Fuel_Type'].unique()
# checking different classes in fuel types of vehicles
```

```python
data['Fuel_Type'] = data['Fuel_Type'].map({'Petrol':0,'Diesel':1,'CNG':2})
# converting categorical values to numeric values for Fuel_type class
```

```python
data['Seller_Type'].unique()
# checking different classes in Seller types of vehicles
```

```python
data['Seller_Type'] = data['Seller_Type'].map({'Dealer':0,'Individual':1})
# converting categorical values to numeric values for Seller_type class
```

```python
data['Transmission'].unique()
data['Transmission'] =data['Transmission'].map({'Manual':0,'Automatic':1})
```

```python
X = data.drop(['Car_Name','Selling_Price', 'Owner'],axis=1)
y = data['Selling_Price']
# Assigning X and Y values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
# cross-validation 80-20
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
#Checking accuracy with 4 different algorithms
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)

rf = RandomForestRegressor()
rf.fit(X_train,y_train)

gb = GradientBoostingRegressor()
gb.fit(X_train,y_train)

xg = XGBRegressor()
xg.fit(X_train,y_train)
#Fitting training set for both the algorithms
```

```python
y_pred1 = lr.predict(X_test)
y_pred2 = rf.predict(X_test)
y_pred3 = gb.predict(X_test)
y_pred4 = xg.predict(X_test)
```

```python
from sklearn import metrics
```

```python
score1 = metrics.r2_score(y_test,y_pred1)
score2 = metrics.r2_score(y_test,y_pred2)
score3 = metrics.r2_score(y_test,y_pred3)
score4 = metrics.r2_score(y_test,y_pred4)
# Saving R2 score
```

```python
print(score1*100, score2*100, score3*100, score4*100)
```
